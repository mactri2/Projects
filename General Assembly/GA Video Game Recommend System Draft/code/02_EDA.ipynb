```python
#creating genres dictionary
g = {
    'key': ['13',
            '24',
            '26',
            '4',
            '15',
            '31',
            '12',
            '5',
            '7',
            '32',
            '16',
            '30',
            '9',
            '11',
            '25',
            '34',
            '8',
            '10',
            '14',
            '33',
            '2',
            'Unknown'],
    'val': ['Simulator',
            'Tactical',
            'Quiz/Trivia',
            'Fighting',
            'Strategy',
            'Adventure',
            'Role-playing (RPG)',
            'Shooter',
            'Music',
            'Indie',
            'Turn-based strategy (TBS)',
            'Pinball',
            'Puzzle',
            'Real Time Strategy (RTS)',
            "Hack and slash/Beat 'em up",
            'Visual Novel',
            'Platform',
            'Racing',
            'Sport',
            'Arcade',
            'Point-and-click',
            'Other']
}
```

```python
print(len(g['key']))
print(len(g['val']))
```

```python
#create genres key value df
cats = pd.DataFrame(g)
cats
```

```python
cats['key'].dtype
```

```python
def map_func(str_keys):
    str_cats = ''
    split_keys = str_keys.split(',')
    for k in split_keys:
        try:
            str_cats += cats[cats['key'] == k]['val'].tolist()[0]
            str_cats += ', '
        except:
            pass
    return str_cats
```

```python
df['genres'].map(lambda x: map_func(x))
```

```python
df['genres'] = df['genres'].map(map_func)
```
